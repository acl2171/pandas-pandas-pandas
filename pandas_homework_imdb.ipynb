```python
# code for Intermediate Task 1 - count how many movies have each of the content ratings
movies['content_rating'].unique()
```

```python
# code for Intermediate Task 2 - use a visualization to display that same data, including a title and x and y labels
movies.groupby('content_rating')['title'].count().plot(kind='bar');
plt.xlabel('Rating')
plt.ylabel("Number of Movies")
plt.title("Movie Ratings");
```

```python
# code for Intermediate Task 3 - convert the following content ratings to "UNRATED": NOT RATED, APPROVED, 
# PASSED, GP
movies=movies.replace({'content_rating':['NOT RATED', 'APPROVED','PASSED','GP']}, 'UNRATED')
```

```python
movies.groupby('content_rating')['title'].count()
```

```python
# code for Intermediate Task 4 - convert the following content ratings to "NC-17": X, TV-MA
movies=movies.replace({'content_rating':['X', 'TV-MA']}, 'NC-17')
```

```python
movies.groupby('content_rating')['title'].count()
```

```python
# code for Intermediate Task 5 - count the number of missing values in each column
movies.isnull().sum()
```

```python
# code for Intermediate Task 6 - if there are missing values: examine them, then fill them in with "reasonable" values
movies['content_rating'].value_counts(dropna=False)
```

```python
movies[movies['content_rating'].isna()]
```

```python
#Using a different datasource (IMDB Parents Guide):
# we find Butch Cassidy as PG. 
# Where Eagles Dare as M which we assume is NC-17. 
# And True Grit as PG-13
movies[movies['title']=='Butch Cassidy and the Sundance Kid'].fillna({'content_rating':'PG'})
```

```python
movies[movies['title']=='Where Eagles Dare'].fillna({'content_rating':'NC-17'})
```

```python
movies[movies['title']=='True Grit'].fillna({'content_rating':'PG-13'})
```
